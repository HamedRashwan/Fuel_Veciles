```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('D:\COURSES\AI\Epsilon Ai\Session 38\Fuel_Veciles\data\processed\datset_feul_cleaned1.csv')
df
```

```python
# Distribution of vehicles by year
plt.figure(figsize=(12, 6))
sns.countplot(x='year', data=df, palette='viridis')
plt.xticks(rotation=90)
plt.title('Distribution of Vehicles by Year')
plt.xlabel('Year')
plt.ylabel('Number of Vehicles')
plt.tight_layout()
plt.show()
```

```python
# Group data by 'Fuel_Type' and 'Engine_Cylinders' and calculate average CO2_In_Grams_Mile
avg_co2_by_fuel_and_cylinders = df.groupby(['fuel_type', 'engine_cylinders'])['tailpipe_co2_in_grams_mile_ft1'].mean().unstack()
```

```python
# Plotting the grouped bar chart
plt.figure(figsize=(12, 10))
avg_co2_by_fuel_and_cylinders.plot(kind='bar', stacked=False)
plt.title('Average tailpipe_co2_in_grams_mile_ft1 by Fuel Type and Cylinder Count')
plt.xlabel('Fuel Type')
plt.ylabel('Average CO2 Emissions (g)')
plt.legend(title='Cylinders', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# Group data by year and calculate the average Combined_MPG
avg_mpg_by_year = df.groupby('year')['combined_mpg_ft1'].mean()
# Plot the trends for average fuel economy over time
plt.figure(figsize=(10, 6))
avg_mpg_by_year.plot(kind='line', marker='o', color='r', linewidth=2)
plt.title('Average Fuel Economy (Combined MPG F1) Over Time')
plt.xlabel('Year')
plt.ylabel('Average Combined MPG')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Distribution of vehicles by make (top 20)
plt.figure(figsize=(12, 6))
top_makes = df['make'].value_counts().nlargest(20)
sns.barplot(x=top_makes.index, y=top_makes.values, palette='viridis')
plt.xticks(rotation=90)
plt.title('Top 20 Vehicle Makes')
plt.xlabel('Make')
plt.ylabel('Number of Vehicles')
plt.tight_layout()
plt.show()
```

```python
# Distribution of fuel types
plt.figure(figsize=(12, 6))
sns.countplot(x='fuel_type', data=df, palette='viridis')
plt.xticks(rotation=45)
plt.title('Distribution of Fuel Types')
plt.xlabel('Fuel Type')
plt.ylabel('Number of Vehicles')
plt.tight_layout()
plt.show()
```

```python
# Distribution of vehicles by class
plt.figure(figsize=(12, 6))
vehicle_classes = df['class'].value_counts()
sns.barplot(x=vehicle_classes.index, y=vehicle_classes.values, palette='viridis')
plt.xticks(rotation=90)
plt.title('Distribution of Vehicles by Class')
plt.xlabel('Class')
plt.ylabel('Number of Vehicles')
plt.tight_layout()
plt.show()
```

```python
# Create a Bar Plot Based on # of Cars Introduced Per Year
ye = df['year'].value_counts() .head(10) .plot(kind='bar', title='Top Years Cars Introduced')
ye.set_xlabel('Year Introduced')
ye.set_ylabel('Count')
```

```python
# Create Histogram Plot for determining the frequency of combined_ft1 fuel economy
cmf1 = df['combined_mpg_ft1'].plot(kind='hist', bins=40, title='Vehicle Fuel Economy', )
cmf1.set_xlabel('Fuel Economy (MPG)')
```

```python
# Create KDE Plot for determining the frequency of combined_1 fuel economy
cmf= df['combined_mpg_ft1'].plot(kind='kde', title='Coaster Speed (mph)')
cmf.set_xlabel('Fuel Economy (MPG)')
```

```python
# Filter data for years 2007-2017
df_ye = df[df['year'].between(2000, 2017)]
# Calculate average fuel economy for each car brand
average_mpg_by_brand_make = df_ye.groupby('make')['combined_mpg_ft1'].mean()

# Sort the car brands based on average fuel economy and select top 10
top_10_brands_make = average_mpg_by_brand_make.sort_values(ascending=False).head(10)

# Plotting the bar chart
plt.figure(figsize=(10, 6))
top_10_brands_make.plot(kind='bar', color='skyblue')
plt.title('Top 10 Car Brands based on Fuel Economy (2000-2017)')
plt.xlabel('Car Brand')
plt.ylabel('Average Combined MPG')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
grouped_by_year_class = df.groupby(['year', 'class']).size().unstack()

# Group the data by year and drivetrain
grouped_by_year_drivetrain = df.groupby(['year', 'drive']).size().unstack()
# Plot the trends for drivetrains
plt.figure(figsize=(12, 10))
grouped_by_year_drivetrain.plot(kind='line', marker='o', linewidth=2)
plt.title('Trends in Drivetrain Popularity Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Vehicles')
plt.legend(title='Drivetrain', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
transmission_data = df[['transmission', 'combined_mpg_ft1']]

# Group data by transmission type and calculate the average Combined_MPG
avg_mpg_by_transmission = transmission_data.groupby('transmission')['combined_mpg_ft1'].mean().sort_values(ascending=False)

# Select the top ten most efficient transmissions
top_ten_transmissions = avg_mpg_by_transmission.head(10)
# Plot the average fuel economy for the top ten most efficient transmission types
plt.figure(figsize=(10, 6))
top_ten_transmissions.plot(kind='bar', color='red')
plt.title('Top Ten Most Efficient Transmissions')
plt.xlabel('Transmission Type')
plt.ylabel('Average Combined MPG')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
